```python
import pygame
import sys

# Inicializar Pygame
pygame.init()

# Configurar pantalla
ANCHO = 800
ALTO = 600
pantalla = pygame.display.set_mode((ANCHO, ALTO))
pygame.display.set_caption("Mi Primer Juego en Pygame")

# Colores (R, G, B)
NEGRO = (0, 0, 0)

# Bucle principal
ejecutando = True
while ejecutando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            ejecutando = False

    # Rellenar fondo
    pantalla.fill(NEGRO)

    # Actualizar pantalla
    pygame.display.flip()

# Salir
pygame.quit()
# sys.exit() # Descomentar en script normal
```

## 2. Dibujar Formas Geométricas

Pygame permite dibujar rectángulos, círculos, líneas y polígonos fácilmente.

```python
import pygame

pygame.init()
pantalla = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Dibujando Formas")

BLANCO = (255, 255, 255)
ROJO = (255, 0, 0)
VERDE = (0, 255, 0)
AZUL = (0, 0, 255)

ejecutando = True
while ejecutando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            ejecutando = False

    pantalla.fill(BLANCO)

    # Dibujar Rectángulo (superficie, color, (x, y, ancho, alto))
    pygame.draw.rect(pantalla, ROJO, (50, 50, 100, 50))

    # Dibujar Círculo (superficie, color, (centro_x, centro_y), radio)
    pygame.draw.circle(pantalla, VERDE, (300, 100), 40)

    # Dibujar Línea (superficie, color, inicio, fin, grosor)
    pygame.draw.line(pantalla, AZUL, (400, 50), (500, 150), 5)

    pygame.display.flip()

pygame.quit()
```

## 3. Movimiento y Animación

Para animar objetos, cambiamos sus coordenadas en cada iteración del bucle principal y controlamos la velocidad de fotogramas (FPS).

```python
import pygame

pygame.init()
pantalla = pygame.display.set_mode((800, 600))
reloj = pygame.time.Clock() # Para controlar FPS

x = 50
y = 300
velocidad_x = 5

ejecutando = True
while ejecutando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            ejecutando = False

    # Lógica de movimiento
    x += velocidad_x
    
    # Rebotar en los bordes
    if x > 800 or x < 0:
        velocidad_x *= -1

    # Dibujar
    pantalla.fill((0, 0, 0))
    pygame.draw.circle(pantalla, (255, 255, 0), (x, y), 30)
    
    pygame.display.flip()
    reloj.tick(60) # Limitar a 60 FPS

pygame.quit()
```

## 4. Control con Teclado

Podemos detectar qué teclas se presionan para controlar objetos.

```python
import pygame

pygame.init()
pantalla = pygame.display.set_mode((800, 600))
reloj = pygame.time.Clock()

x = 400
y = 300
velocidad = 5

ejecutando = True
while ejecutando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            ejecutando = False

    # Obtener estado de todas las teclas
    teclas = pygame.key.get_pressed()
    
    if teclas[pygame.K_LEFT]:
        x -= velocidad
    if teclas[pygame.K_RIGHT]:
        x += velocidad
    if teclas[pygame.K_UP]:
        y -= velocidad
    if teclas[pygame.K_DOWN]:
        y += velocidad

    pantalla.fill((0, 0, 0))
    pygame.draw.rect(pantalla, (0, 255, 255), (x, y, 50, 50))
    
    pygame.display.flip()
    reloj.tick(60)

pygame.quit()
```

## 5. Manejo del Mouse

Podemos obtener la posición del mouse y detectar clics.

```python
import pygame

pygame.init()
pantalla = pygame.display.set_mode((800, 600))

ejecutando = True
while ejecutando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            ejecutando = False
        
        # Detectar clic
        if evento.type == pygame.MOUSEBUTTONDOWN:
            if evento.button == 1: # Clic izquierdo
                print("Clic izquierdo en:", evento.pos)

    # Obtener posición del mouse
    pos_mouse = pygame.mouse.get_pos()

    pantalla.fill((255, 255, 255))
    
    # Dibujar círculo donde está el mouse
    pygame.draw.circle(pantalla, (255, 0, 255), pos_mouse, 20)
    
    pygame.display.flip()

pygame.quit()
```
